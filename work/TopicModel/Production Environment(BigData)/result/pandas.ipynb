```python
#############################################
import numpy as np
csv = pd.read_csv('topic05brands.csv',encoding="utf-8")
df=pd.DataFrame(csv)
# print(df)
df2=df.sort_index(ascending=True)
a=list(df2.values.flatten())
Xb=[]
Yb=[]
for i,aa in enumerate(a):
    if(i%2==0):
        Xb.append(aa)
    else:
        Yb.append(aa)
#print(Xb)

Yb=np.array(Yb)
Yb/=sum(Yb)
#print(Yb)
Yb=list(Yb) #リストに戻す
```

```python
csv = pd.read_csv('topic05tags.csv',encoding="utf-8")
df=pd.DataFrame(csv)
# print(df)
df2=df.sort_index(ascending=True)
a=list(df2.values.flatten())
Xt=[]
Yt=[]
for i,aa in enumerate(a):
    if(i%2==0):
        Xt.append(aa)
    else:
        Yt.append(aa)
#print(Xt)

Yt=np.array(Yt)
Yt/=sum(Yt)
#print(Yt)
Yt=list(Yt) #リストに戻す
```

```python
df3 = pd.DataFrame(columns=['brand','rate1','tag','rate2'])
df3['brand']=Xb[0:10]
df3['rate1']=Yb[0:10]
df3['tag']=Xt[0:10]
df3['rate2']=Yt[0:10]
```
