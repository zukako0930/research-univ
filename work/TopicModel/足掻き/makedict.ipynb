```python
with open('brands.csv','r',encoding='utf8') as fbrand:
    txt=fbrand.read() 
    #print(txt)
    lines = txt.split("\n")
    Coordinate=[]
    #重複のない辞書にするためにsetを用いる
    s=set()
    for line in lines:
        brands=line.split(",")
        Coordinate.append(brands)
        for brand in brands:
            s.add(brand)
    brand1=list(s)
    print(len(brand1))
    print(brand1)
```

```python
#dictionary型に変換する関数
def two_list_to_dict(key_list, val_list):
      return dict(zip(key_list, val_list))
```

```python
index=[]
for i in range(len(brand1)):
    index.append(i)
print(index)
```
