```python
brands=pd.read_csv("./0.1_0.02_0.02_1/topic42brands.csv")
tags=pd.read_csv("./0.1_0.02_0.02_1/topic42tags.csv")

brands=brands.rename(columns={'0':'brand','1':'counts'})
brands.counts=brands.counts/brands.counts.sum()

tags=tags.rename(columns={'0':'tag','1':'counts'})
tags.counts=tags.counts/tags.counts.sum()
```

```python
noise_tags=pd.read_csv("./0.1_0.02_0.02_1/topic42noise_tags.csv")
noise_tags=noise_tags.rename(columns={'0':'tag','1':'counts'})
# noise_tags.counts=noise_tags.counts/noise_tags.counts.sum()
```

```python
#ノイズとして現れたタグのリストを作る=50トピックすべてをconcatして.tagをsetにする．
noise_df_list=[]
for i in range(50):
    if  i<10:
        noise_tags=pd.read_csv("./0.1_0.02_0.02_1/topic0"+str(i)+"noise_tags.csv")
        noise_tags=noise_tags.rename(columns={'0':'tag','1':'counts'})   
    else:
        noise_tags=pd.read_csv("./0.1_0.02_0.02_1/topic"+str(i)+"noise_tags.csv")
        noise_tags=noise_tags.rename(columns={'0':'tag','1':'counts'})
    noise_df_list.append(noise_tags)
```

```python
new=noise_df_list[0]
for i in range(49):
    new=pd.concat([new,noise_df_list[i]])
```

```python
noise_tag_set=list(set(new.tag))
```

```python
noise_tag_count=np.zeros(len(noise_tag_set))
for i in range(len(new)):
    noise_tag_count[noise_tag_set.index(new.iloc[i].tag)]+=new.iloc[i].counts
```

```python
a=pd.DataFrame({'tag':noise_tag_set})
b=pd.DataFrame({'counts':noise_tag_count})
ab=pd.concat([a,b],axis=1)
```

```python
ab=ab.sort_values(by=['counts'],ascending=False)
```
