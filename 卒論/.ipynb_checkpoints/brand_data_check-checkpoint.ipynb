### 出現頻度で絞ろう

```python
import pandas as pd
import numpy as np
```

```python
import os,csv
with open('tags_120000.csv') as f:
    data = csv.reader(f)
#     print(data)
    tags_list=[]
    for dat in data:
        for da in dat:
            tags_list.append(da)
#     print(brands_list)
with open('brands_120000.csv') as f:
    data = csv.reader(f)
#     print(data)
    brands_list=[]
    for dat in data:
        for da in dat:
            brands_list.append(da)
#     print(brands_list)
```

```python
brands_set=list(set(brands_list))
```

```python
print(len(brands_set))
```

```python
def count_tags(data_list,data_set):
    count_vec=np.ndarray(len(data_set))
    for data in data_list:
#         print(data)
        count_vec[data_set.index(data)]+=1
#     print(count_vec)
    return count_vec
```

```python
count_vec=count_tags(brands_list,brands_set)
```

```python
max(count_vec)
```

```python
count_vec
```

```python
import matplotlib.pyplot as plt
% %matplotlib inline
```

```python
left=np.ndarray(len(count_vec))
plt.bar(left,count_vec)
```

### 出現頻度の高いタグと低すぎるタグを確認して削除する

```python
new_list=[]
count_list=list(count_vec)
new_list=[count_list.index(x) for x in count_list if x > 100 and x<1000] 
```
