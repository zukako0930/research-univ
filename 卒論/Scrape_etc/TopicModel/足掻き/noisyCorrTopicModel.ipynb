```python
#dictionary型に変換する関数
def two_list_to_dict(key_list, val_list):
      return dict(zip(key_list, val_list))
```

```python
#ブランドの辞書作成
with open('./data/brands.csv','r',encoding='utf8') as fbrand:
    txt=fbrand.read() 
    #print(txt)
    lines = txt.split("\n")
    Coordinate=[]
    #重複のない辞書にするためにsetを用いる
    s=set()
    for line in lines:
        brands=line.split(",")
        Coordinate.append(brands)
        for brand in brands:
            s.add(brand)
    brand1=list(s)
    print(len(brand1))
    #print(brand1)   
#dictionaryのindex用
index=[]
for i in range(len(brand1)):
    index.append(i)
#print(index)   
#順番よくわからんけど辞書できた？
brand_dict=two_list_to_dict(brand1,index)
print(brand_dict)
```

```python
#タグの辞書作成
#出現回数の少ないタグは無視していいかも
with open('./data/tags.csv','r',encoding='utf8') as ftag:
    txt=ftag.read() 
    lines = txt.split("\n")
    tags_list=[]
    #重複のない辞書にするためにsetを用いる
    s=set()
    for line in lines:
        tags=line.split(",")
        tags_list.append(tags)
        for tag in tags:
            s.add(tag)
    alltags=list(s)
    print(len(alltags))
    #print(alltags)
#dictionaryのindex用
index=[]
for i in range(len(alltags)):
    index.append(i)
#辞書作成
tag_dict=two_list_to_dict(alltags,index)
print(tag_dict)
```

```python
with open('./data/brands.csv','r',encoding='utf8') as fbrand:
    txt=fbrand.read() 
    lines = txt.split("\n")
    Coordinates=[]
    for line in lines:
        brands=line.split(",")
        Coordinates.append(brands)
    #print(Coordinates)
    
W = [] # words
X = [] # categories
#ブランドと補助情報の文書ごとのリストを作る
for Coordinate in Coordinates:
    #list()をつけない場合iterator objectが帰ってくる
    W.append(list(map(lambda v:brand_dict[v],Coordinate)))
```

```python
# データ生成 
#dictionary型
vocab = {'computer':0, 'banana':1, 'ipad':2, 'orange':3, 'apple':4}
categories = {'FOODS':0, 'TECH':1, 'TOREAD':2}

Vw = len(vocab)
Vx = len(categories)

W = [] # words
X = [] # categories

#lambda v:vocab[v]が第一引数, それをmap関数により['computer','ipad','apple']に順番に適用
#vにはvocabのindexが返される
W.append(map(lambda v:vocab[v], ['computer', 'ipad', 'apple']))
X.append(map(lambda c:categories[c], ['TECH', 'TOREAD']))
W.append(map(lambda v:vocab[v], ['ipad', 'ipad']))
X.append(map(lambda c:categories[c], ['TECH', 'TOREAD']))
W.append(map(lambda v:vocab[v], ['ipad', 'ipad', 'apple', 'apple']))
X.append(map(lambda c:categories[c], ['TOREAD']))
W.append(map(lambda v:vocab[v], ['banana', 'orange', 'apple']))
X.append(map(lambda c:categories[c], ['FOODS', 'TOREAD']))
W.append(map(lambda v:vocab[v], ['banana', 'orange']))
X.append(map(lambda c:categories[c], ['FOODS']))
```

```python
print(W)
print(X)
```

```python
# モデル定義
K = 2
alpha = 0.1
beta= 0.1
gamma= 0.1
eta = 1.0
max_iter = 100
nctm = NCTM(K=K, alpha=alpha, beta=beta, gamma=gamma, eta=eta, max_iter=max_iter)
```

```python
# NCTM のフィッティング
nctm.fit(W,X,Vw,Vx)
```
