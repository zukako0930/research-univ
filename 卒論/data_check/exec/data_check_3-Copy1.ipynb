````python
    #dictionaryåž‹ã«å¤‰æ›ã™ã‚‹é–¢æ•°
    def two_list_to_dict(key_list, val_list):
          return dict(zip(key_list, val_list))
    
    #csvã‚’èª­ã¿è¾¼ã‚“ã§è¡Œã”ã¨ã«åˆ†å‰²
    def read_file(filename):
        with open(str(filename),'r',encoding='utf8') as fbrand:
            txt=fbrand.read() 
            lines = txt.split("\n")
            return lines
        
    #csvã®å„è¡Œã‚’åˆ†å‰²ã—ã¦ã‚³ãƒ¼ãƒ‡ã”ã¨ã®ãƒ–ãƒ©ãƒ³ãƒ‰listã®listã‚’ä½œã‚‹.
    def make_Coordinate(lines):
        Coordinate=[]
        #é‡è¤‡ã®ãªã„è¾žæ›¸ã«ã™ã‚‹ãŸã‚ã«setã‚’ç”¨ã„ã‚‹
        s=set()
        for line in lines:
            brands=line.split(",")
            Coordinate.append(brands)
        return Coordinate
    
    #good_tagsã®ã¿ã«çµžã‚‹
    def make_goodtags(lines,good_tags_set):
        good_tags_list=[]
        #é‡è¤‡ã®ãªã„è¾žæ›¸ã«ã™ã‚‹ãŸã‚ã«setã‚’ç”¨ã„ã‚‹
        s=set()
        s=good_tags_set
        for line in lines:
            tags=line.split(",")
            good_tags=[t for t in tags if t in s] 
            #if len(good_tags) == 0:
            good_tags_list.append(good_tags)
        return good_tags_list
    
    def make_brandset(lines):
        #é‡è¤‡ã®ãªã„è¾žæ›¸ã«ã™ã‚‹ãŸã‚ã«setã‚’ç”¨ã„ã‚‹
            s=set()
            for line in lines:
                brands=line.split(",")
                for brand in brands:
                    s.add(brand)
            brand_set=list(s)
            print(len(brand_set))
            return brand_set
    
    def make_dict(brandset):
        #dictionaryã®index
        index=[]
        for i in range(len(brandset)):
            index.append(i)
        #è¾žæ›¸
        brand_dict=two_list_to_dict(brandset,index)
        return brand_dict
    
    def brand_to_dictnum(Coordinates,brand_dict):
        W = [] # words
        #ãƒ–ãƒ©ãƒ³ãƒ‰ã¨è£œåŠ©æƒ…å ±ã®æ–‡æ›¸ã”ã¨ã®ãƒªã‚¹ãƒˆã‚’ä½œã‚‹
        for Coordinate in Coordinates:
        #list()ã‚’ã¤ã‘ãªã„å ´åˆiterator objectãŒå¸°ã£ã¦ãã‚‹
            W.append(list(map(lambda v:brand_dict[v],Coordinate)))
        return W
    
    #çµæžœã«é–¢ã™ã‚‹é–¢æ•°  
    def brand_to_topic(Coordinates,nctm_Z): #nctm.Zã¯nctm.fitã®çµæžœãŒå…¥ã£ã¦ã„ã‚‹
        brand_topic=[]
        for brands, Topics in zip(Coordinates,nctm_Z):
            for brand,Topic in zip(brands,Topics):
                pair=[brand,Topic]
                brand_topic.append(pair)
        return brand_topic
    
    def brand_topic_list_f(brand_topic,K):
        brand_topic_list=[]
        for i in range(K):
            aaa=[x for x in brand_topic if x[1] == i]
            #print(aaa)
            brand_topic_list.append(aaa)
        return brand_topic_list
    
    def t_setlist_brand_f(brand_topic_list,K):
        #brandã«ã¤ã„ã¦
        t_setlist_brand=[] #å„ãƒˆãƒ”ãƒƒã‚¯ã”ã¨ã®brandã®setã‚’ä»£å…¥ã™ã‚‹ãƒªã‚¹ãƒˆ
        for i in range(K): #å„ãƒˆãƒ”ãƒƒã‚¯ã§å›žã™
            a=[] #å„ãƒˆãƒ”ãƒƒã‚¯ã§setã‚’ä½œã‚‹ãŸã‚ã®ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°
            for item in brand_topic_list[i]:
                a.append(item[0])
            a=set(a) #ã‚»ãƒƒãƒˆã«ã™ã‚‹
            a=list(a) #ãƒªã‚¹ãƒˆã«ç›´ã™
            t_setlist_brand.append(a) #ç¾åœ¨ã®ãƒˆãƒ”ãƒƒã‚¯ã®setã‚’ã‚°ãƒ­ãƒ¼ãƒãƒ«ã®ãƒªã‚¹ãƒˆã«è¿½åŠ 
        #print(t_setlist[0]) 
        return t_setlist_brand
    
    def count_topic_f(brand_topic_list,t_setlist_brand,K):
        count_topic=[] #å…¨ãƒˆãƒ”ãƒƒã‚¯ã®å‡ºç¾å›žæ•°ãƒªã‚¹ãƒˆã®ãƒªã‚¹ãƒˆ
        for i in range(K): #ãƒˆãƒ”ãƒƒã‚¯ãƒ«ãƒ¼ãƒ—
            count=np.zeros((len(t_setlist_brand[i]))) #ä»Šè¦‹ã¦ã„ã‚‹ãƒˆãƒ”ãƒƒã‚¯ã®setã®é•·ã•åˆ†ã®brandã‚«ã‚¦ãƒ³ãƒˆé…åˆ—ã‚’åˆæœŸåŒ–
            for item in brand_topic_list[i]: #itemã¯ãƒªã‚¹ãƒˆ
                count[t_setlist_brand[i].index(item[0])]+=1 #ä»Šè¦‹ã¦ã„ã‚‹brandã®ã‚¤ãƒ³ãƒ‡ãƒƒã‚¯ã‚¹ã‚’ã‚«ã‚¦ãƒ³ãƒˆã‚¢ãƒƒãƒ—
            count_topic.append(count) #ä¸€ã¤ã®ãƒˆãƒ”ãƒƒã‚¯ã«ã¤ã„ã¦è¦‹çµ‚ãˆãŸã‚‰ã‚°ãƒ­ãƒ¼ãƒãƒ«ã«è¿½åŠ 
        return count_topic
    
    def write_result(filename,t_setlist_brand,count_topic):
        topic_num=0
        for item,count in zip(t_setlist_brand,count_topic):
            tuples=[]
            for b,c in zip(item,count):
                tuples.append(tuple((b,c)))
            tuples.sort(key=lambda x:x[1],reverse=True) #x[1]ã«å‡ºç¾å›žæ•°ãŒå…¥ã£ã¦ã„ã‚‹ã®ã§å¤šã„é †ã«ã‚½ãƒ¼ãƒˆ
            df=pd.DataFrame(tuples)
            if topic_num < 10:
                df.to_csv( 'result/topic'+'0'+str(topic_num)+str(filename), index=False )
            else:
                df.to_csv( 'result/topic'+str(topic_num)+str(filename), index=False )
            topic_num+=1
        print("END")
````
