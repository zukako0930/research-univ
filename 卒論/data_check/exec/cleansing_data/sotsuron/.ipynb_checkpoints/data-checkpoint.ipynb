```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
files=os.listdir()
```

```python
files
```

```python
perps=[]
for f in files:
    with open(f,'r') as p:
        reader = csv.reader(p)
        for row in reader:
            perp=np.array(row)
            perp=perp.astype(float)
#             print(type(perp))
        perps.append(perp.min())
```
