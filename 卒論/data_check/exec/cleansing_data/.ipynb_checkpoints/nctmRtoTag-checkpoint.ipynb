## nctmRをロードしてタグ情報と紐づけるためのプログラムです．

```python
def read_file(filename):
    with open(str(filename),'r',encoding='utf8') as fbrand:
        txt=fbrand.read() 
        lines = txt.split("\n")
        return lines
```

```python
def make_Coordinate(lines):
    Coordinate=[]
    #重複のない辞書にするためにsetを用いる
    s=set()
    for line in lines:
        brands=line.split(",")
        Coordinate.append(brands)
    return Coordinate
```

```python
nctmR=make_Coordinate(read_file('./result_iter1000/topic100/pattern001/nctmR.csv'))
```

```python
print(nctmR[0:10])
```

```python
tags=make_Coordinate(read_file('./goodtags10000.csv'))
```

```python
print(tags[0:10])
```

```python
def noise_judge(tags,nctmR):
    noise_tags=[] #r==0 tag
    signif_tags=[] #r==1 tag 
    notnoise_tags=[]
    for t_list,r_list in zip(tags,nctmR):
        for t,r in zip(t_list,r_list):
#             print(t,r)
            if r == '1':
                signif_tags.append(t)
            else:
                noise_tags.append(t)
    #setにする？
    return signif_tags,noise_tags
```

```python
signif_tags,noise_tags = noise_judge(tags,nctmR)
```

```python
print(signif_tags[0:100])
```
