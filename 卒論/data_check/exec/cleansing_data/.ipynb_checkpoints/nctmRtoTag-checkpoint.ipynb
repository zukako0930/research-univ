## nctmRをロードしてタグ情報と紐づけるためのプログラムです．

```python
def read_file(filename):
    with open(str(filename),'r',encoding='utf8') as fbrand:
        txt=fbrand.read() 
        lines = txt.split("\n")
        return lines
```

```python
def make_Coordinate(lines):
    Coordinate=[]
    #重複のない辞書にするためにsetを用いる
    s=set()
    for line in lines:
        brands=line.split(",")
        Coordinate.append(brands)
    return Coordinate
```

```python
nctmR=make_Coordinate(read_file('./result_iter10000/topic100/pattern001/nctmR.csv'))
```

```python
print(nctmR[0:10])
```

```python
tags=make_Coordinate(read_file('./goodtags10000.csv'))
```

```python
print(tags[0:10])
```

```python
def noise_judge(tags,nctmR):
    noise_tags=[] #r==0 tag
    signif_tags=[] #r==1 tag 
    notnoise_tags=[]
    for t_list,r_list in zip(tags,nctmR):
        for t,r in zip(t_list,r_list):
#             print(t,r)
            if r == '1':
                signif_tags.append(t)
            else:
                noise_tags.append(t)
    #setにする？
    return signif_tags,noise_tags
```

```python
signif_tags,noise_tags = noise_judge(tags,nctmR)
```

```python
print(signif_tags[0:100])
```

```python
# tagのsetを作って0と判定された回数をカウントしたい
def get_tagset(tags):
    xxx_list=[]
    for ts in tags:
        for t in ts:
            xxx_list.append(t)
    return list(set(xxx_list))
```

```python
tagset=get_tagset(tags)
len(tagset)
```

```python
def get_countvec(tagset,noise_tags):
    import numpy as np
    count_vec=np.zeros(len(tagset))
    for s_t in noise_tags:
        count_vec[tagset.index(s_t)]+=1
    return count_vec
```

```python
count=get_countvec(tagset,noise_tags)
```

```python
def get_countvec_fromlist(tagset,tags):
    import numpy as np
    count_vec=np.zeros(len(tagset))
    for ts in tags:
        for t in ts: 
            count_vec[tagset.index(t)]+=1
    return count_vec
```

```python
tag_emcount=get_countvec_fromlist(tagset,tags)
```

```python
te=list(tag_emcount)
print(max(tag_emcount))
te.index(max(tag_emcount))
```
