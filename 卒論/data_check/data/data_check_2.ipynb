```python
# csvèª­ã¿è¾¼ã¿ã®ãŸã‚ã®é–¢æ•° brandã¯æ°—ã«ã—ãªã„ã§
def read_csvfile(filename):
    with open(filename,'r',encoding='utf8') as f:
        txt=f.read()
        lines=txt.split('\n')
    brands_list=[]
    for line in lines:
#         brands=[]
#         for l in lines:
#             brands.append(l)
        brands_list.append(line)
    return brands_list
```

```python
#å‡ºç¾å›žæ•°ã‚«ã‚¦ãƒ³ãƒˆç”¨é–¢æ•°
def emerge_count(data_list,data_set):
    count_vec=np.ndarray(len(data_set))
    for data in data_list:
#         print(data)
        count_vec[data_set.index(data)]+=1
#     print(count_vec)
    return count_vec
```
