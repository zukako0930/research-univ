# データの前処理を行う
## 1.データの読み込み

```python
import numpy as np
```

```python
# csv読み込みのための関数 brandは気にしないで
def read_csvfile(filename):
    with open(filename,'r',encoding='utf8') as f:
        txt=f.read()
        lines=txt.split('\n')
    brands_list=[]
    for line in lines:
#         brands=[]
#         for l in lines:
#             brands.append(l)
        brands_list.append(line)
    return brands_list
```

```python
#brands=read_csvfile('brands.csv')
tags=read_csvfile('../data/tags.csv')
```

```python
tags[1:10]
```

```python
#tagのsetを作るために一度全てを同じlistに追加する
t0=[]
for tag in tags:
    t0.append(tag.split(','))
```

```python
print(t0[0:10])
```

```python
t1=[]
for s0 in t0:
    for s1 in s0:
        if s1 != '':
            t1.append(s1)
```

```python
## 一次元のリストになっているか確認
print(t1[0:10])
```

```python
##  setを作る
tags_set=list(set(t1))
# tags_set.remove('')
tags_set
```

```python
# t1は一次元のlistなのでこれを元にカウントすれば良い
tags_list=t1
```

```python
#出現回数カウント用関数
def emerge_count(data_list,data_set):
    count_vec=np.ndarray(len(data_set))
    for data in data_list:
#         print(data)
        count_vec[data_set.index(data)]+=1
#     print(count_vec)
    return count_vec
```

```python
tags_emerge_count=emerge_count(tags_list,tags_set)
```

```python
tags_emerge_count[0:10]
```

```python
print(max(tags_emerge_count))
a=list(tags_emerge_count)
print(a.index(17156.0))
print(tags_set[a.index(17156.0)])
```

### 冬がめちゃめちゃ多いっぽい
`tags_emerge_count` : タグの出現回数を格納したlist
`tags_set` : `tags_emerge_count`と同じ長さで出現回数1回以上のtagが格納されているlist
`tags_emerge_count`と`tags_set`はindexで紐づけられている．
`t1` : 全てのタグを一次元のlistにflattenしたもの

#### これからやること
- `tags_emerge_count`から出現回数一定数以上(以下)のindexのlistを作る⇨`good_tag_idxs`
    - この時`tags_emerge_count`の中に重複がないことを確認する．⇨~~重複あるやろ普通~~
- `good_tag_idxs`を利用して`tags_set`からそれらの文字列のlistを作る⇨`good_tags_set`
- 元データ`t0`をfor文で読み込みながら，`good_tags_set`の中に存在しない文字列は`remove`してしまう．
-　この操作によってtagが0になったindexのデータは，`brands.csv`から読み込んだデータからも削除して整合性を保つ．
    - 「タグがない」データが存在しないようにする．


```python
good_tags_idxs=[]
##　一回リストに追加したidxはgood_tags_idxsに追加しないようにする．

# for count in count_list:
#     if count > 
# good_tags_idxs=[count_list.index(x) for x in count_list if x > 10] 
```

```python
## count_listの中で出現回数が10回以上のものをgood_tags_idxsに入れる
good_tags_idxs=[]
count_list=list(tags_emerge_count)
## 出現回数一定数以上のタグのidxをlistにする．
good_tags_idxs = [i for i, x in enumerate(count_list) if x > 10]
print(len(good_tags_idxs))
```

```python
print(good_tags_idxs)
```

```python
## good_tags_idxsに基づいてタグを絞る
good_tags=[]
for idx in good_tags_idxs:
    good_tags.append(tags_set[idx])
print(good_tags)
```

## 出現回数が一定以上のタグだけに絞ることはできた．
これを元に入れるデータを作る．
- brands.csvはそのまま
- tags.csvの中でgood_tagsに入っている個数２以上のデータを無視して新たなデータとする．
- この時tags.csvの中でgood_tagsに入っていないtagはそもそも無視する．
- さらに，brands_setに含まれるtagも(ブランド名と一致するので)消去していいかもしれない．
    - 逆にこれがトピックとタグが一致している教師データになりうる．
