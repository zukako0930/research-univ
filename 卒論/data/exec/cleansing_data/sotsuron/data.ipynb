```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
files=os.listdir('./topic_perp/')
```

```python
files[1]
```

```python
perps=[]
for f in files:
    with open(''f,'r') as p:
        reader = csv.reader(p)
        for row in reader:
            perp=np.array(row)
            perp=perp.astype(float)
#             print(type(perp))
        perps.append(perp.min())
```

```python
perps=[]
```

```python
perp_list=[]
for i in range(15):
    with open('./topic_perp/Perplexity_topic'+str((i+1)*10)+'-Copy1.csv','r') as p:
            reader = csv.reader(p)
            for row in reader:
                perp=np.array(row)
                perp=perp.astype(float)
            perp_list.append(perp)
```

```python
perp_list=np.array(perp_list)
```

```python
c=['-r','-b','-g','darkviolet','cyan','deeppink','black','brown','darkblue','darkgray']
x=np.linspace(1,1000,1000)
for i in range(10):
    plt.plot(x,perp_list[i],c[i],label='topic'+str((i+1)*10))
plt.xlabel("iterations")
plt.ylabel("perplexity")
# plt.ticklabel_format(style='sci',axis='y',scilimits=(0,0))
plt.legend()
plt.savefig('10-100_perp.png')
plt.show()    
```
